Load packages

```python
import pandas as pd
```

Read data

```python
mtcars
```

## Linear regression with `statsmodels`

```python
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
model_linear = smf.ols('mpg ~ disp', data = mtcars).fit()
```

```python
model_linear.summary()
```
